```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
```

```python
csv_file = './Resources/Cleaned_Saferparks.csv'

saferparks = pd.read_csv(csv_file, sep = '\t', index_col = None)
saferparks
```

```python
saferparks_grouped = saferparks.groupby('Category')
saferparks_grouped.mean()
```

```python
saferparks["Year"] = saferparks["Date"].map(lambda x: x.split("-")[2])
saferparks['Month'] =  [x.split('-')[1] for x in saferparks.loc[:,'Date']]
saferparks['Day'] =  [int(x.split('-')[0]) for x in saferparks.loc[:,'Date']]
saferparks['Week'] = [1+int(x)//7 for x in saferparks.loc[:,'Day']]
```

```python
test=saferparks.groupby("Year")
count=test["Injury"].count()
years=saferparks["Year"].unique()
years
```
